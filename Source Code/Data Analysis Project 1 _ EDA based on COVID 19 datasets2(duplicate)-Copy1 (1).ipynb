# Vizualise the data

bargraphs with active cases, recovered cases, etc

Country_Region against Confirmed cases

```python
top_10_confirmed = csv_file_t.nlargest(10, 'Confirmed')

print("Top 10 countries with the most confirmed cases:")
print(top_10_confirmed[['Country_Region', 'Confirmed']])

#Bar plot
colormap = plt.get_cmap('plasma') 
top_10_confirmed.plot(kind='bar', x='Country_Region', y='Confirmed', legend=False, title='Top 10 countries with most confirmed cases', color=colormap(np.linspace(0, 1, len(top_10_confirmed))))
plt.ticklabel_format(style='plain', axis='y')
plt.show()

#Pie plot
top_10_confirmed.plot.pie(y='Confirmed', labels=top_10_confirmed['Country_Region'], autopct='%1.1f%%', legend=False, title='Confirmed Cases Distribution')
plt.ylabel('')
plt.show()
```

```python
top_10_confirmed = csv_file_t.nlargest(10, 'Deaths')

print("Top 10 countries with the most deaths cases:")
print(top_10_confirmed[['Country_Region', 'Deaths']])

#Bar plot
colormap = plt.get_cmap('plasma') 
top_10_confirmed.plot(kind='bar', x='Country_Region', y='Deaths', legend=False, title='Top 10 countries with most deaths cases', color=colormap(np.linspace(0, 1, len(top_10_confirmed))))
plt.ticklabel_format(style='plain', axis='y')
plt.show()

#Pie plot
top_10_confirmed.plot.pie(y='Deaths', labels=top_10_confirmed['Country_Region'], autopct='%1.1f%%', legend=False, title='Deaths Cases Distribution')
plt.ylabel('')
plt.show()
```

```python
top_10_confirmed = csv_file_t.nlargest(10, 'Recovered')

print("Top 10 countries with the most recovered cases:")
print(top_10_confirmed[['Country_Region', 'Recovered']])

#Bar plot
colormap = plt.get_cmap('plasma') 
top_10_confirmed.plot(kind='bar', x='Country_Region', y='Recovered', legend=False, title='Top 10 countries with most recovered cases', color=colormap(np.linspace(0, 1, len(top_10_confirmed))))
plt.ticklabel_format(style='plain', axis='y')
plt.show()

#Pie plot
top_10_confirmed.plot.pie(y='Recovered', labels=top_10_confirmed['Country_Region'], autopct='%1.1f%%', legend=False, title='Recovered Cases Distribution')
plt.ylabel('')
plt.show()
```

```python
top_10_confirmed = csv_file_t.nlargest(10, 'Active')

print("Top 10 countries with the most active cases:")
print(top_10_confirmed[['Country_Region', 'Active']])

#Bar plot
colormap = plt.get_cmap('plasma') 
top_10_confirmed.plot(kind='bar', x='Country_Region', y='Active', legend=False, title='Top 10 countries with most Active cases', color=colormap(np.linspace(0, 1, len(top_10_confirmed))))
plt.ticklabel_format(style='plain', axis='y')
plt.show()

#Pie plot
top_10_confirmed.plot.pie(y='Active', labels=top_10_confirmed['Country_Region'], autopct='%1.1f%%', legend=False, title='Active Cases Distribution')
plt.ylabel('')
plt.show()
```
